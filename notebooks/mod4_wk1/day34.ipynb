### Part 1a: enumerate() Fluency (Day 3/3 - Final Day)

```python
# Task data
api_responses = [
    {"status": 200, "data": "success"},
    {"status": 404, "data": None},
    {"status": 500, "data": "server error"},
    {"status": 200, "data": "success"}
]

# Task: Loop through with enumerate(), print which responses failed (status != 200)
# Include the index position in your output
# Expected output format:
# Response 1 failed with status 404
# Response 2 failed with status 500

for i, response in enumerate(api_responses):
    status = response.get("status")
    
    if status != 200:
        print(f"Response {i} failed with status {status}")
```

### Part 1b: .get() Pattern (Day 3/5)

**Reinforcing the None vs Missing distinction from Day 33:**

```python
# Pattern reminder (corrected version):
# .get(key, default) returns default ONLY when key is MISSING
# If key exists with value None, returns None (not default)

# Solution for nested dicts with possible None:
# (outer_dict.get('key') or {}).get('nested_key', 'default')

# Task data - API response with inconsistent structure
properties = [
    {"address": {"city": "Austin", "zip": "78701"}, "price": 450000},
    {"address": None, "price": 380000},  # address exists but is None
    {"price": 290000},  # address key missing entirely
    {"address": {"city": None, "zip": "78205"}, "price": 520000}  # city is None
]

# Task: Extract city from each property, defaulting to "Unknown" when unavailable
# Use the (dict.get('key') or {}).get('nested', 'default') pattern
# Expected output:
# Property 0: Austin
# Property 1: Unknown
# Property 2: Unknown
# Property 3: Unknown

for i, prop in enumerate(properties, start=1):
    city = (prop.get("address") or {}).get("city") or "Unknown"
    print(f"Property {i}: {city}")
```
