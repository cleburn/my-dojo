```python
num_list = list(range(1,21))
print(num_list)
```

```python
# print only even numbers, 2 ways to do it
print(num_list[1::2])

# or
print([num for num in num_list if num%2==0])
```

```python
odd_sq_list = []
for num in num_list:
    if num % 2 == 1:
        odd_sq_list.append(num**2)
    else:
        pass
print(odd_sq_list)
```

```python
new_list_sum = 0
for num in odd_sq_list:
    new_list_sum += num
print(new_list_sum)
```

```python
def mean_return(x):
    mean = 0
    total = 0
    for num in x:
        total += num
    mean += total/len(x)
    return mean

print(mean_return(num_list))
print(mean_return(odd_sq_list))
```

```python
def larger_num(x,y):
    if x > y:
        return x
    else:
        return y

print(larger_num(47, 582))
```

```python
def above_threshold(x, y):
    count = 0
    for num in x:
        count += 1 if num > y else 0
    return count

print(above_threshold(num_list, 8))
```
