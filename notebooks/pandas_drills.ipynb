```python
df['price'] = df['price'].replace({'\$':'',',':''}, regex=True).astype(float)
df[df['price']>150]
```

```python
df[(df['price']>150) & (df['bedrooms']>=2)]
```

```python
df[['name','price','host_neighbourhood']]
```
