```python
import pandas as pd
import numpy as np
```

```python
data = [10,12,15,18,20,22,25,100]
df = pd.Series(data)
print("Mean:",df.mean())
print("Median:", df.median())
print("Std:", df.std())
```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=8, edgecolor='black')
plt.axvline(df.mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df.median(), color='blue', linestyle='--', label='Median')
plt.legend()
plt.show()
```
