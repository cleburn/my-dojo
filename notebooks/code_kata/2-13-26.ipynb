```python
competitors = ["Alice", "Bob", "Charlie"]

for i,name in enumerate(competitors, start=1):
    print(f"{i}. {name}")
```

```python
names = ["Item A", "Item B"]
prices = [10, 20]

for item,price in zip(names,prices):
    print(f"{item} costs ${price}")
```

```python
files = ["report.pdf", "image.png", "data.csv", "photo.jpg"]

images = [f for f in files if f.endswith((".png", ".jpg"))]

print(images)
```

```python
required = {"flour", "sugar", "eggs", "butter"}
have = {"flour", "butter"}

missing = required - have

print(missing)

print(required.difference(have))
```

```python
values = ["10", "20", "oops", "30"]

for i in values:
    try:
        num = int(i)
        print(num * 2)
    except ValueError:
        print("Bad data")
```

```python
matrix = [[1, 2], [3, 4]]

flat = [num for row in matrix for num in row]
print(flat)
```

```python
word = "radar"

is_palindrome = word == word[::-1]
print(is_palindrome)
```
