```python
squares = [num**2 for num in range(1,21) if num%2==0]
print(squares)

big_squares = [num**2 for num in range(1,21) if (num%2==0) & (num**2>50)]
print(big_squares)
```

```python
transactions = [
    ("Alice", 150),
    ("Bob", 200),
    ("Alice", 75),
    ("Charlie", 300),
    ("Bob", 100)
]

total_per_person = {}
for name,val in transactions:
    total_per_person[name] = total_per_person.get(name, 0) + val

print(total_per_person)
```

```python
import pandas as pd

df = pd.DataFrame({
    'price': [100, 200, None, 150, None, 300],
    'bedrooms': [2, 3, 3, None, 4, 2],
    'sqft': [1200, 1500, 1400, None, 1800, 1300]
})

print("Missing Count:")
print(df.isnull().sum())

print("\nMissing %:")
print((df.isnull().mean() * 100).round(2))

row_missing = df.isnull().sum(axis=1)
max_missing = row_missing.max()
bad_rows = df[row_missing == max_missing]

print(f"\nRow(s) with most missing values({max_missing}):")
print(bad_rows)
```
