```python
walkers = {
    "Dad": [42, 'male'],
    "Clebo": [16, 'male'],
    "Liam": [14, 'male'],
    "Henry": [11, 'male']
}

ages = [info[0] for info in walkers.values()]

youngest = min(ages)
oldest = max(ages)

print(f"The walkers range in age from {youngest} to {oldest}")
```

```python
family = {
    "Dad": [42, 'male'],
    "Clebo": [16, 'male'],
    "Liam": [14, 'male'],
    "Henry": [11, 'male']
}

family["Delta"] = [21, 'female']
family["Brooke"] = [20, 'female']

age = [age for name, (age, gender) in family.items()]
name = [name for name, (age, gender) in family.items()]
gender = [gender for name, (age, gender) in family.items()]

print(f"The Walkers range in age from {min(age)} to {max(age)}")
print(f"\nThe family has {gender.count('male')} boys and {gender.count('female')} girls")
print(f"\nTheir names are {name}")
```
