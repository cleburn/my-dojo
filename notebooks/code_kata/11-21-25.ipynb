```python
words = """Here in about an hour I'm going to go pick up Kelly and Brenda 
to go to the airport so they can pick up a rental car. Then they are gonna drive it to California."""
```

```python
def mega_paragraph_breakdown(x,y,z):
    return print(
        f"Total Word Count: {
        len(x.split())
        }\n\nShouting Version: {
        x.upper()
        }\n\nChanging 'Brenda' to 'her Mom': {
        x.replace(y,z)
        }")

mega_paragraph_breakdown(words,'Brenda','her Mom')
```

```python
next_words = """I have been a dad for 21 years, 
and I will never really know if I've done a good enough job. 
But I try my best every day, in every way. And hopefully when the kids are grown
they will think of me with grace and respect and joy."""

mega_paragraph_breakdown(next_words,'dad','father')
```
