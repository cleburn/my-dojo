# Dictionaries

```python
people = {"Dad": 42, "Delta": 21, "Brooke": 20, "Clebo": 16, "Liam": 14}
print(people)
```

```python
people["Henry"]=11.5
print(people)
```

```python
avg = sum(people.values()) / len(people.values())
print(f"The overall average Walker age is: {avg:.1f}")
print("Walkers above average age:")
for key, value in people.items():
    if value >= avg:
        print(f"{key}")
```

# Statistics Practice

```python
import numpy as np

dataset = [14,17,18,34,55,101,2035]
df = pd.Series(dataset)
print(f"The mean is: {df.mean():.2f}")
print(f"The median is: {df.median()}")
print(f"The variance is: {df.var():.2f}")
print(f"The standard deviation is: {df.std():.2f}")
```
