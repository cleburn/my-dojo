```python
def even_odd(x):
    if x % 2 == 0:
        return print("even")
    else:
        return print("odd")
even_odd(34)
even_odd(49)
even_odd(9047365892)
```

```python
def div_by(x):
    if (x % 3 == 0) and (x % 5 == 0):
        return print("divisible by 3 & 5")
    elif x % 5 == 0:
        return print("divisible by 5")
    elif x % 3 == 0:
        return print("divisible by 3")
    else:
        return print("not divisible by 3 or 5")

div_by(9)
div_by(15)
div_by(25)
div_by(49)
```
