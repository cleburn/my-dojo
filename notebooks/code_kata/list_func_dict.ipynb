```python
def even_odd(x):
    if x % 2 == 0:
        return print("even")
    else:
        return print("odd")

even_odd(48)
even_odd(49)
```

```python
walker_fam = {"Dad": 42, "Delta": 21, "Brooke": 20, "Clebo": 16, "Liam": 14, "Henry": 12}
avg_walker_age = sum(walker_fam.values()) / len(walker_fam.values())
print(f"Average Walker age: {avg_walker_age:.2f}")
```

```python
print(f'Walkers above average age:')
for key, val in walker_fam.items():
    if val >= avg_walker_age:
        print(f'{key}')
```
