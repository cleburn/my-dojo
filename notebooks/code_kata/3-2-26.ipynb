```python
import pandas as pd
import numpy as np

print("All imports successful")
```

```python
df_sales = pd.DataFrame({
    "Employee": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Sales": [450, 800, 300, 950, 600]
})

top_sales = df_sales[df_sales["Sales"] > 500].sort_values("Sales", ascending=False)
print(top_sales)
```

```python
df_scores = pd.DataFrame({
    "Student": ["Tom", "Jerry", "Spike", "Tyke"],
    "Score": [85, np.nan, 92, np.nan]
})

df_scores["Score"] = df_scores["Score"].fillna(df_scores["Score"].mean())
print(df_scores)
```

```python
df_products = pd.DataFrame({
    "Category": ["Tech", "Clothing", "Tech", "Food", "Clothing", "Tech", "Tech"]
})

proportions = df_products['Category'].value_counts(normalize=True)
print(proportions)

percentages = (df_products['Category'].value_counts(normalize=True) * 100).round(2)
print(f"\n{percentages}")
```

```python
df_expenses = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "Marketing", "HR", "IT"],
    "Spend": [1200, 500, 800, 2000, 300, 1500]
})

df_expenses.groupby("Department")["Spend"].sum()
```

```python
df_finance = pd.DataFrame({
    "Product": ["Widget A", "Widget B", "Widget C"],
    "Revenue": [100, 250, 80],
    "Cost": [60, 150, 70]
})

df_finance["Profit"] = df_finance["Revenue"] - df_finance["Cost"]
print(df_finance)
```
