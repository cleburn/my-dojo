```python
products = [
    {"name": "Keyboard", "price": 49.99},
    {"name": "Mouse", "price": 24.99},
    {"name": "USB Cable", "price": 12.99},
    {"name": "Monitor", "price": 199.99}
]

cheap = next((p for p in products if p["price"] < 20), None)
print(cheap)
```

```python
users = [
    {"name": "Alice", "verified": True},
    {"name": "Bob", "verified": False},
    {"name": "Carol"}
]

for i in users:
    if i.get("verified") == True:
        print(f"Name: {i.get("name")} (verified)")
    else:
        print(f"Name: {i.get("name")} (unverified)")
```
