# Functions

```python
# - Write a function that takes a list and returns max, min, and range
# - Add a parameter that lets you choose whether to round the result

def thur_fun(x, do_round=False):
    if do_round is False:
        return print(f"List Max: {max(x)}\nList Min: {min(x)}\nList Range: {max(x) - min(x)}")
    else:
        return print(f"List Max: {round(max(x))}\nList Min: {round(min(x))}\nList Range: {round(max(x) - min(x))}")

practice = [7,8,9,2,14,55,73,102,4,2,2,81]

thur_fun(practice)
```

```python
new = [ 1,3,5,7,99,32,57,0,21,5,15,18,83,64,1002]
thur_fun(new)
```

```python
round_list = [14.85, 5.87, 5.56, 3.14]

thur_fun(round_list)
thur_fun(round_list, do_round=True)
```
