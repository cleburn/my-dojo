```python
bids = [427, 157, 112, 479, 240, 225, 214, 171, 477, 152]

high_bids = [bid for bid in bids if bid > 300]
print(f"There were {len(high_bids)} bids over $300")

closest_to_250 = max(bid for bid in bids if bid < 250)
print(f"The highest bid under $250 was {closest_to_250}")
```

```python
daily_sales = [2, 5, 3, 8, 1, 4, 6, 2, 9, 7]

current_streak = 1
longest_streak = 1

current_start = 0
longest_start = 0

for i in range(1, len(daily_sales)):
    if daily_sales[i] > daily_sales[i-1]:
        current_streak +=1
        if current_streak > longest_streak:
            longest_streak = current_streak
            longest_start = current_start
    else:
        current_streak = 1
        current_start = i

longest_end = longest_start + longest_streak

print(f"Longest streak: {longest_streak} days")
print(f"From index {longest_start} to {longest_end-1}")
print(f"Values: {daily_sales[longest_start:longest_end]}")
```

```python
temperatures = [65, 68, 70, 67, 71, 73, 75, 72, 74, 76, 78]

temp_streak = 1
longest_temp_streak = 1

temp_start = 0
longest_temp_start = 0

for i in range(1, len(temperatures)):
    if temperatures[i] > temperatures[i-1]:
        temp_streak += 1
        if temp_streak > longest_temp_streak:
            longest_temp_streak = temp_streak
            longest_temp_start = temp_start
    else:
        temp_streak = 1
        temp_start = i

longest_temp_end = longest_temp_start + longest_temp_streak

print(f"Longest streak: {longest_temp_streak} days")
print(f"From index: {longest_temp_start} to {longest_temp_end-1}")
print(f"Values: {temperatures[longest_temp_start:longest_temp_end]}")
```

```python
cabinet = {
    "width": 37,
    "height": 72,
    "length": 24
}

w, h, l = cabinet.values()
volume = w * h * l
surface_area = 2 * (w*h + w*l + h*l)
perimeter = sum(cabinet.values())

print(volume)
print(surface_area)
print(perimeter)
```

```python
budget = 100
expenses = [25, 10, 50, 5, 20]

if sum(expenses) > budget:
    print(f"Over budget by ${sum(expenses)-budget}")
else:
    print(f"Under budget by ${budget - sum(expenses)}")
```
