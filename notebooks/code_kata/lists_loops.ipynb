```python
num_list = list(range(1,16))
print(num_list)
```

```python
for num in num_list:
    print(num)
```

```python
# print only evens and odds, using 2 separate methods

print(num_list[1::2])

print([num for num in num_list if num%2==1])
```

```python
# Sum & Mean manually

list_sum = 0

for num in num_list:
    list_sum += num

print(f"The sum of the list is: {list_sum}\nThe mean of the list is: {list_sum / len(num_list)}")
```

```python
# Sum & Mean with Built-ins

print(f"The list sum is: {sum(num_list)}")
print(f"The list mean is: {sum(num_list)/len(num_list)}")
print(f"The largest number in the list is: {max(num_list)}")
print(f"The smalles number in the list is: {min(num_list)}")
print(f"There are {len(num_list)} numbers  in the list")
```
