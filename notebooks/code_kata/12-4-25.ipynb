```python
walker_fam = {
    "Dad": ["Male", 42],
    "Delta": ["Female", 21],
    "Brooke": ["Female", 20],
    "Clebo": ["Male", 16],
    "Liam": ["Male", 14],
    "Henry": ["Male", 11]
}
print(walker_fam)
```

```python
# 1. Print the entire dict
print("Full dict:")
print(walker_fam)
print()

# 2. Check each person individually
print("Individual values:")
for name, info in walker_fam.items():
    print(f"{name}: {info}, Type: {type(info)}, Length: {len(info)}")
print()

# 3. Run the comprehension
comp_ages = [val[1] for val in walker_fam.values()]
print("Ages from comprehension:")
print(comp_ages)
print()

# 4. Check types in result
print("Types in comp_ages:")
for i, age in enumerate(comp_ages):
    print(f"Index {i}: {age}, Type: {type(age)}")
```

```python
avg_age = sum(ages)/len(ages)
print(avg_age)
```

```python
old_walkers = []
young_walkers = []

for key,val in walker_fam.items():
    if val[1] >= avg_age:
        old_walkers.append(key)
    else:
        young_walkers.append(key)

print(f"""The average age in the Walker family is {
avg_age:.2f}.\nPeople over the average age are {
old_walkers}. And everyone under the average age is {young_walkers}.""")

# Another way to structure the print statement that works and looks cleaner is:
#   print(f"The average age in the Walker family is {avg_age:.2f}.\n"
#        f"People over the average age are {old_walkers}. "
#        f"And everyone under the average age is {young_walkers}.")
```

```python
Walkers = {
    "Dad": ["Male", 42],
    "Delta": ["Female", 21],
    "Brooke": ["Female", 20],
    "Clebo": ["Male", 16],
    "Liam": ["Male", 14],
    "Henry": ["Male", 11]
}
print(Walkers)
```

```python
print(Walkers['Brooke'])
```

```python
boys = [key for key,val in Walkers.items() if val[0] == "Male"]
girls = [key for key,val in Walkers.items() if val[0] == "Female"]

print(f"The Walker family has {len(boys)} boys named {boys}\n"
      f"And the Walker family has {len(girls)} girls named {girls}")
```

```python
Walker_ages = [(key,val[1]) for key,val in Walkers.items()]
print(Walker_ages)
```

```python
just_ages = [val[1] for val in Walkers.values()]
average_walker_age = round(sum(just_ages) / len(just_ages))

print(just_ages)
print(average_walker_age)
```

```python
below_average_boys = [key for key,val in Walkers.items() if (val[1] < average_walker_age) & (val[0] == "Male")]
above_average_girls = [key for key,val in Walkers.items() if (val[1] > average_walker_age) & (val[0] == "Female")]

print(f"The average Walker age is: {average_walker_age}.\n"
      f"The Walker boys younger than average are: {below_average_boys}.\n"
      f"The Walker girls over the average age is/are: {above_average_girls}.\n"
      f"There are {len(below_average_boys)} boys under the average age, and\n"
      f"There are {len(above_average_girls)} girls over the average age."
     )
```
