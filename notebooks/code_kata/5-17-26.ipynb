```python
phrase = "Version Control is essential for engineers."
vowels = "aeiouAEIOU"

vow_count = 0
for i in phrase:
    if i in vowels:
        vow_count += 1
print(vow_count)

vow_count_2 = sum(1 for char in phrase if char in vowels)

print(vow_count_2)
```

```python
front_end = {"HTML": 90, "CSS": 85}
back_end = {"Python": 95, "SQL": 80}

full_stack = {}
for key,val in front_end.items() and back_end.items():
    if key not in full_stack:
        full_stack[key] = val
print(full_stack)

# correct Merge Operator 
full_stack_2 = front_end | back_end
print(full_stack_2)
```
