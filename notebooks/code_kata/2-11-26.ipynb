```python
bids_rand = [random.randint(100,500) for _ in range(10)]

bids = [102, 479, 312, 202, 445, 224, 252, 492, 458, 276]

high_count = sum(bid > 300 for bid in bids)
print(f"Bids over $300: {high_count}")

under_250 = [bid for bid in bids if bid <= 250]
if under_250:
    closest = max(under_250)
    print(f"Closest to $250: ${closest}")
else:
    print("No bids under $250")
```

```python
daily_sales = [2, 5, 3, 8, 1, 4, 6, 2, 9, 7]

current_streak = 1
longest_streak = 1

for i in range(1, len(daily_sales)):
    if daily_sales[i] > daily_sales[i-1]:
        current_streak += 1
        longest_streak = max(longest_streak, current_streak)
    else:
        current_streak = 1

print(f"Longest increasing streak: {longest_streak} days")
```

```python
sales = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']


fruit_count = {}
for fruit in sales:
    fruit_count[fruit] = fruit_count.get(fruit, 0) + 1

print(fruit_count)
```
