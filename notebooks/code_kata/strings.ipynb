```python
import pandas as pd

df = pd.read_csv('../../data/raw/listings.csv')
df.shape
```

```python
df.info()
```

```python
pd.set_option('display.max_rows',None)
df.head(1).T
```

```python
df['price'] = df['price'].str.replace('[^\d.]','',regex=True).astype(float)
df.head(3)
```

```python
df.info()
```

```python
narrowed_df = df.groupby('listing_url')[['estimated_occupancy_l365d','estimated_revenue_l365d']].max().head(20)
narrowed_df.sort_values('estimated_revenue_l365d',ascending=False).head(20)
```

```python
narrowed_df.loc[['https://www.airbnb.com/rooms/1000016109686428185','https://www.airbnb.com/rooms/1002881780954180678']]
```

```python
top_2 = narrowed_df.sort_values('estimated_revenue_l365d', ascending=False).head(2)
top_2_urls = top_2.index

full_data = df[df['listing_url'].isin(top_2_urls)]
pd.set_option('display.max_columns',None)
full_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

print(f"There is a {df['neighbourhood_cleansed'].corr(df['price']):.2f} correlation between ZIP Code and Price")
print('='*55)

plt.scatter(data=df,x='neighbourhood_cleansed',y='price', alpha=0.4)

plt.xlabel('ZIP Codes')
plt.ylabel('Estimated Revenue ($)')
plt.title('Price based on area, or Outliers?')
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
plt.savefig('../../visuals/codekata_scatter.png')
plt.show()
```
