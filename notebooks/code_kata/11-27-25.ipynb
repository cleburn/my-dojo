```python
import numpy as np
```

```python
nums = list(range(1,11))
print(nums)
```

```python
even_nums = nums[1::2]
odd_nums = nums[::2]
print(even_nums)
print(odd_nums)
```

```python
even_squared = []
for num in even_nums:
    even_squared.append(num**2)

odd_squared = []
for num in odd_nums:
    odd_squared.append(num**2)

print(even_squared)
print(odd_squared)
```

```python
array = np.array(odd_squared)
print(array)
```

```python
print(array.mean())
print(round(array.std(),3)) # rounding method
print(f"{array.std():.3f}") # f-string method
```

```python
sentence = """It's Turkey Day, and I'm being lazy. But I'm about to get in the kitchen to make breakfast,
then get the turkey in the oven. Once that's going, I am headed to the gym with Clebo. Then back home to eat and be 
lazy with the boys. Movies, poker, and later maybe get some work done on my book."""

def sent_structure(x):
    print(f"Word Count: {len(list(x))}")
    print('='*30)
    print(f"Whisper Version:\n{x.lower()}")
    print('='*30)
    print(f"Shouting Version:\n{x.upper()}")

sent_structure(sentence)
```
